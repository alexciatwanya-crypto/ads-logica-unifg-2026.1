```python
class Estudante:
    def __init__(self, nome, matricula):
        self.nome = nome
        self.matricula = matricula
        self.notas = []

    def adicionar_nota(self, nota):
        if not isinstance(nota, (int, float)):
            raise ValueError("A nota deve ser um número.")
        if not (0 <= nota <= 10):
            raise ValueError("A nota deve estar entre 0 e 10.")
        self.notas.append(nota)

    def calcular_media(self):
        if not self.notas:
            return 0
        return sum(self.notas) / len(self.notas)

    def situacao(self):
        media = self.calcular_media()
        if media >= 7:
            return "Aprovado"
        else:
            return "Recuperação"

    def __str__(self):
        return f"Estudante: {self.nome} (Matrícula: {self.matricula})"


# Demonstração da classe Estudante
estudante1 = Estudante("Alice", "12345")
estudante2 = Estudante("Bob", "67890")

print(f"Estudante 1: {estudante1.nome}, Matrícula: {estudante1.matricula}, Notas: {estudante1.notas}")
print(f"Estudante 2: {estudante2.nome}, Matrícula: {estudante2.matricula}, Notas: {estudante2.notas}")

try:
    estudante1.adicionar_nota(8.5)
    estudante1.adicionar_nota(9.0)
    estudante1.adicionar_nota(6.0)
    print(f"{estudante1.nome} - Notas após adição: {estudante1.notas}")
    print(f"{estudante1.nome} - Média: {estudante1.calcular_media():.2f}")
    print(f"{estudante1.nome} - Situação: {estudante1.situacao()}")

    estudante2.adicionar_nota(5.0)
    estudante2.adicionar_nota(6.5)
    print(f"{estudante2.nome} - Notas após adição: {estudante2.notas}")
    print(f"{estudante2.nome} - Média: {estudante2.calcular_media():.2f}")
    print(f"{estudante2.nome} - Situação: {estudante2.situacao()}")
except ValueError as e:
    print(f"Erro ao adicionar nota: {e}")


class Produto:
    imposto_padrao = 0.10

    def __init__(self, nome, preco):
        self.nome = nome
        self.preco = preco

    def preco_com_imposto(self):
        return self.preco * (1 + Produto.imposto_padrao)


# Demonstração da classe Produto
produto1 = Produto("Livro", 50.00)
produto2 = Produto("Caneta", 5.00)

print(f"\nProduto: {produto1.nome}, Preço: R${produto1.preco:.2f}, Preço com imposto: R${produto1.preco_com_imposto():.2f}")
print(f"Produto: {produto2.nome}, Preço: R${produto2.preco:.2f}, Preço com imposto: R${produto2.preco_com_imposto():.2f}")

Produto.imposto_padrao = 0.15
print(f"Imposto padrão alterado para 15%:\nProduto: {produto1.nome}, Preço: R${produto1.preco:.2f}, Preço com imposto: R${produto1.preco_com_imposto():.2f}")


class Turma:
    def __init__(self, nome_turma):
        self.nome_turma = nome_turma
        self.estudantes = []

    def matricular(self, estudante):
        if not isinstance(estudante, Estudante):
            raise TypeError("Apenas objetos Estudante podem ser matriculados.")
        self.estudantes.append(estudante)
        print(f"{estudante.nome} matriculado(a) na turma {self.nome_turma}.")

    def media_geral(self):
        if not self.estudantes:
            return 0
        total_medias = 0
        contador = 0
        for estudante in self.estudantes:
            if estudante.notas:
                total_medias += estudante.calcular_media()
                contador += 1
        return total_medias / contador if contador > 0 else 0

    def gerar_relatorio(self):
        print(f"\n--- Relatório da Turma: {self.nome_turma} ---")
        if not self.estudantes:
            print("Nenhum estudante matriculado nesta turma.")
            return
        for estudante in self.estudantes:
            media = estudante.calcular_media()
            situacao = estudante.situacao()
            print(f"  - {estudante.nome} (Matrícula: {estudante.matricula}) - Média: {media:.2f} - Situação: {situacao}")
        print(f"Média Geral da Turma: {self.media_geral():.2f}")
        print("-------------------------------------")


# Demonstração da classe Turma
turma_a = Turma("Turma A - Python")
turma_a.matricular(estudante1)
turma_a.matricular(estudante2)

estudante3 = Estudante("Carlos", "11223")
estudante3.adicionar_nota(7.0)
estudante3.adicionar_nota(7.5)
turma_a.matricular(estudante3)

turma_a.gerar_relatorio()


class ExemploSelfErro:
    def __init__(self, valor):
        self.valor = valor

    def imprimir_valor_correto(self):
        print(f"\nO valor (correto) é: {self.valor}")

obj_erro = ExemploSelfErro(100)
obj_erro.imprimir_valor_correto()


class ExemploSemSelf:
    def __init__(self, dado):
        self.dado = dado

    def tenta_acessar_dado(self):
        print(f"Acessando o dado da instância: {self.dado}")

obj_sem_self = ExemploSemSelf("dado importante")
obj_sem_self.tenta_acessar_dado()


# Reflexão Final: Atributos de Classe vs. Atributos de Instância
class Animal:
    especie = "Mamífero"

    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade

    def __str__(self):
        return f"{self.nome} ({self.idade} anos) - Espécie: {self.especie}"


animal1 = Animal("Leão", 5)
animal2 = Animal("Tigre", 3)

print(f"\nAnimal 1: {animal1}")
print(f"Animal 2: {animal2}")

print(f"Nome do animal1: {animal1.nome}")
print(f"Nome do animal2: {animal2.nome}")

print(f"Espécie do animal1: {animal1.especie}")
print(f"Espécie do animal2: {animal2.especie}")
print(f"Espécie diretamente da classe: {Animal.especie}")

animal1.nome = "Leão Dourado"
print(f"\nNome do animal1 após modificação: {animal1.nome}")
print(f"Nome do animal2 permanece: {animal2.nome}")

Animal.especie = "Felino Grande"

print(f"\nEspécie do animal1 após modificação na classe: {animal1.especie}")
print(f"Espécie do animal2 após modificação na classe: {animal2.especie}")
print(f"Espécie diretamente da classe: {Animal.especie}")

animal1.especie = "Gato Doméstico (erro)"
print(f"\nAnimal 1 com 'erro' de espécie: {animal1.especie} (agora é um atributo de instância!)")
print(f"Animal 2 ainda é da espécie da classe: {animal2.especie}")
print(f"Classe Animal ainda tem a espécie: {Animal.especie}")

print("\nConclusão:")
print("Atributos de classe são definidos na classe e compartilhados. Alterar o atributo na classe afeta todas as instâncias.")
print("Atributos de instância são definidos dentro do `__init__` (geralmente) e são únicos para cada objeto. Alterar um atributo de instância não afeta outros objetos ou o atributo de classe.")
print("Se um atributo de classe é sobrescrito em uma instância, um novo atributo de instância com o mesmo nome é criado, e a instância passa a usar seu próprio atributo em vez do de classe.")
```
