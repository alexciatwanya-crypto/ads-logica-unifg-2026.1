```python
import datetime

print("\n--- Exercício 1: O Registro do Sistema ---")

nome_usuario = input("Digite o nome do usuário: ")
ano_nascimento_str = input("Digite o ano de nascimento: ")
altura_str = input("Digite a altura em metros (ex: 1.75): ")

try:
    ano_nascimento = int(ano_nascimento_str)
    altura = float(altura_str)
except ValueError:
    print("Erro: Certifique-se de digitar números válidos para ano de nascimento e altura.")
else:
    ano_atual = datetime.datetime.now().year
    idade = ano_atual - ano_nascimento
    print(f"Olá, {nome_usuario}! Você tem {idade} anos e sua altura é de {altura:.2f}m. Registro concluído.")

print("\n--- Exercício 2: Calculadora de Freelancer ---")
# Solicita o valor por hora e as horas estimadas
valor_hora_str = input("Digite o valor cobrado por hora (ex: 50.00): ")
horas_estimadas_str = input("Digite a estimativa de horas para conclusão (ex: 40): ")

try:
    valor_hora = float(valor_hora_str)
    horas_estimadas = float(horas_estimadas_str)
except ValueError:
    print("Erro: Certifique-se de digitar números válidos para valor por hora e horas estimadas.")
else:
    # Calcula os valores
    valor_bruto = horas_estimadas * valor_hora
    impostos = valor_bruto * 0.15
    valor_liquido = valor_bruto - impostos

    # Exibe os resultados
    print(f"Valor Bruto do Projeto: R$ {valor_bruto:.2f}")
    print(f"Valor dos Impostos (15%): R$ {impostos:.2f}")
    print(f"Valor Líquido Final: R$ {valor_liquido:.2f}")

print("\n--- Exercício 3: Divisão Justa ---")
# Solicita o número total de fatias e de programadores
total_fatias_str = input("Digite o número total de fatias de pizza: ")
num_programadores_str = input("Digite o número de programadores na equipe: ")

try:
    total_fatias = int(total_fatias_str)
    num_programadores = int(num_programadores_str)
except ValueError:
    print("Erro: Certifique-se de digitar números inteiros válidos.")
else:
    if num_programadores == 0:
        print("Erro: Não é possível dividir por zero programadores.")
    else:
        # Calcula fatias por pessoa e o que sobra
        fatias_por_pessoa = total_fatias // num_programadores
        sobra_fatias = total_fatias % num_programadores

        # Exibe os resultados
        print(f"Cada programador comerá {fatias_por_pessoa} fatia(s) inteira(s).")
        print(f"Sobrarão {sobra_fatias} fatia(s) na caixa.")

print("\n--- Exercício 4: Verificador de Acesso ---")
# Solicita a idade e anos de experiência
idade_str = input("Digite sua idade: ")
anos_experiencia_str = input("Digite seus anos de experiência: ")

try:
    idade = int(idade_str)
    anos_experiencia = int(anos_experiencia_str)
except ValueError:
    print("Erro: Certifique-se de digitar números inteiros válidos.")
else:
    # Verifica a condição de acesso sem if/else
    acesso_liberado = (idade >= 18) and (anos_experiencia > 2)

    # Exibe o resultado
    print(f"Acesso Liberado: {acesso_liberado}")

print("\n--- Exercício 5: Desafio do Download ---")
# Solicita o tamanho do arquivo e a velocidade da internet
tamanho_mb_str = input("Digite o tamanho do arquivo em Megabytes (MB): ")
velocidade_mbps_str = input("Digite a velocidade da internet em Megabits por segundo (Mbps): ")

try:
    tamanho_mb = float(tamanho_mb_str)
    velocidade_mbps = float(velocidade_mbps_str)
except ValueError:
    print("Erro: Certifique-se de digitar números válidos para tamanho do arquivo e velocidade.")
else:
    if velocidade_mbps == 0:
        print("Erro: A velocidade da internet não pode ser zero.")
    else:
        # Converte Mbps para MB/s (1 Byte = 8 bits)
        velocidade_mb_por_s = velocidade_mbps / 8

        # Calcula o tempo total em segundos
        tempo_segundos_total = tamanho_mb / velocidade_mb_por_s

        # Converte para minutos e segundos restantes
        minutos_inteiros = int(tempo_segundos_total // 60)
        segundos_restantes = int(tempo_segundos_total % 60)

        # Exibe o resultado
        print(f"{minutos_inteiros} minutos e {segundos_restantes} segundos.")
```
