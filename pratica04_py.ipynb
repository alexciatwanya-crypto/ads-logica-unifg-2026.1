```python
# --- Exercício 1: Limpeza e padronização de nomes ---
print("\n--- Exercício 1: Limpeza e padronização de nomes ---")
nomes_brutos = ["  ana", "BRUNO  ", "cArLa silva", "  joão pedro  "]
nomes_padronizados = []
for nome in nomes_brutos:
    nomes_padronizados.append(nome.strip().title())
print(f"Nomes padronizados: {nomes_padronizados}")
print(f"Total de nomes na lista: {len(nomes_padronizados)}")

# --- Exercício 2: Separando e reunindo texto ---
print("\n--- Exercício 2: Separando e reunindo texto ---")
nome_completo = "Maria Clara Souza"
partes_nome = nome_completo.split()
print(f"Lista com as partes do nome: {partes_nome}")
nome_hifenizado = "-".join(partes_nome)
print(f"Nome unido por hífen: {nome_hifenizado}")
# Desafio opcional
primeiro_nome = partes_nome[0]
ultimo_sobrenome = partes_nome[-1]
print(f"Primeiro nome: {primeiro_nome}, Último sobrenome: {ultimo_sobrenome}")

# --- Exercício 3: Indexação em strings e listas ---
print("\n--- Exercício 3: Indexação em strings e listas ---")
palavra = "algoritmo"
notas = [7.0, 8.5, 6.0, 9.0, 7.5]
print(f"Primeira letra da palavra: {palavra[0]}")
print(f"Quarta letra da palavra: {palavra[3]}")
print(f"Primeira nota da lista: {notas[0]}")
print(f"Última nota da lista: {notas[-1]}")
# Desafio opcional: Em Python (e muitas outras linguagens), o primeiro índice é 0 (zero-based indexing) porque
# representa o deslocamento a partir do início da sequência. O primeiro elemento está a 0 posições do início.

# --- Exercício 4: Fatiamento de sequências ---
print("\n--- Exercício 4: Fatiamento de sequências ---")
palavra = "programacao"
valores = [10, 20, 30, 40, 50, 60]
print(f"Os 4 primeiros caracteres da palavra: {palavra[:4]}")
print(f"Caracteres da posição 4 até a 8: {palavra[4:9]}") # Fatiamento é exclusivo no final
print(f"Os três primeiros elementos da lista: {valores[:3]}")
print(f"Elementos da posição 2 até o final da lista: {valores[2:]}")
# Desafio opcional: Slicing permite especificar início, fim (exclusivo) e passo. Ex: palavra[::2] pega caracteres alternados.

# --- Exercício 5: Percurso simples com for ---
print("\n--- Exercício 5: Percurso simples com for ---")
frutas = ["maçã", "banana", "uva", "pera"]
contador_frutas = 0
for fruta in frutas:
    print(f"Eu gosto de {fruta}.")
    contador_frutas += 1
print(f"Quantidade de frutas percorridas: {contador_frutas}")
# Desafio opcional
palavra_python = "Python"
for letra in palavra_python:
    print(f"Letra: {letra}")

# --- Exercício 6: Percurso por índice ---
print("\n--- Exercício 6: Percurso por índice ---")
alunos = ["Ana", "Bruno", "Carla", "Daniel"]
for i in range(len(alunos)):
    print(f"Índice {i} -> {alunos[i]}")

# --- Exercício 7: Busca linear de um elemento ---
print("\n--- Exercício 7: Busca linear de um elemento ---")
estudantes = ["Ana", "Bruno", "Carla", "Daniel"]
procurado = "Carla"
encontrado = False
for estudante in estudantes:
    if estudante == procurado:
        encontrado = True
        break
if encontrado:
    print(f"O estudante '{procurado}' foi encontrado na lista.")
else:
    print(f"O estudante '{procurado}' NÃO foi encontrado na lista.")

# Desafio opcional: Testando com um nome inexistente
procurado_inexistente = "Pedro"
encontrado_inexistente = False
for estudante in estudantes:
    if estudante == procurado_inexistente:
        encontrado_inexistente = True
        break
if encontrado_inexistente:
    print(f"O estudante '{procurado_inexistente}' foi encontrado na lista.")
else:
    print(f"O estudante '{procurado_inexistente}' NÃO foi encontrado na lista.")

# --- Exercício 8: Contagem de ocorrências ---
print("\n--- Exercício 8: Contagem de ocorrências ---")
itens = ["mouse", "teclado", "mouse", "monitor", "mouse", "teclado"]
contagem_mouse = itens.count("mouse")
contagem_teclado = itens.count("teclado")
print(f"'mouse' aparece {contagem_mouse} vezes.")
print(f"'teclado' aparece {contagem_teclado} vezes.")
# Desafio opcional
contagem_monitor = itens.count("monitor")
print(f"'monitor' aparece {contagem_monitor} vezes.")

# --- Exercício 9: Construção incremental de listas ---
print("\n--- Exercício 9: Construção incremental de listas ---")
tarefas_academicas = []
tarefas_academicas.append("Estudar Python")
tarefas_academicas.append("Resolver exercícios")
tarefas_academicas.append("Revisar código")
tarefas_academicas.append("Enviar atividade")
print(f"Lista final de tarefas: {tarefas_academicas}")
print(f"Total de tarefas cadastradas: {len(tarefas_academicas)}")
# Desafio opcional
tarefas_academicas.append("Preparar apresentação")
print(f"Lista com 5 tarefas: {tarefas_academicas}")

# --- Exercício 10: Filtragem de notas ---
print("\n--- Exercício 10: Filtragem de notas ---")
notas = [4.5, 7.0, 8.0, 5.5, 9.0, 6.8, 7.2]
aprovados = []
for nota in notas:
    if nota >= 7.0:
        aprovados.append(nota)
print(f"Notas aprovadas: {aprovados}")
print(f"Quantidade de estudantes aprovados: {len(aprovados)}")

# --- Exercício 11: Cadastro simples de presença ---
print("\n--- Exercício 11: Cadastro simples de presença ---")
presentes_bruto = ["  maria ", "JOÃO", "ana clara", "Bruno  ", "  carla"]
consulta_bruta = "joão"

nomes_padronizados_presenca = []
for nome in presentes_bruto:
    nomes_padronizados_presenca.append(nome.strip().lower())

consulta_padronizada = consulta_bruta.strip().lower()

print(f"Lista de nomes padronizados: {nomes_padronizados_presenca}")
if consulta_padronizada in nomes_padronizados_presenca:
    print(f"O estudante '{consulta_bruta}' ESTÁ presente.")
else:
    print(f"O estudante '{consulta_bruta}' NÃO está presente.")

# Desafio opcional: Testar com outro nome
consulta_inexistente = "pedro"
consulta_inexistente_padronizada = consulta_inexistente.strip().lower()
if consulta_inexistente_padronizada in nomes_padronizados_presenca:
    print(f"O estudante '{consulta_inexistente}' ESTÁ presente.")
else:
    print(f"O estudante '{consulta_inexistente}' NÃO está presente.")

# --- Exercício 12: Modelagem inicial com coleções lineares ---
print("\n--- Exercício 12: Modelagem inicial com coleções lineares ---")
# (a) nome completo de um aluno -> string
nome_aluno = "Alice Smith" # É uma sequência de caracteres coesa.

# (b) notas de quatro avaliações -> lista
notas_avaliacoes = [8.5, 7.0, 9.2, 6.8] # É uma coleção de itens (notas) que podem ser iterados e acessados individualmente.

# (c) comentário escrito pelo professor -> string
comentario_professor = "Excelente trabalho, continue assim!" # É um bloco de texto.

# (d) tarefas de um projeto -> lista
tarefas_projeto = ["Definir escopo", "Criar cronograma", "Implementar funcionalidade"] # É uma coleção ordenada de itens (tarefas).

# (e) código de matrícula -> string
codigo_matricula = "2023001234" # Geralmente é um identificador que pode conter letras e números, manipulado como texto.

print(f"Nome do aluno (string): {nome_aluno}")
print(f"Notas das avaliações (lista): {notas_avaliacoes}")
print(f"Comentário do professor (string): {comentario_professor}")
print(f"Tarefas do projeto (lista): {tarefas_projeto}")
print(f"Código de matrícula (string): {codigo_matricula}")

# --- Exercício 13: Erros comuns e correção ---
print("\n--- Exercício 13: Erros comuns e correção ---")
# a) lista = [3, 1, 2]
#    resultado = lista.sort()
#    print(resultado)
# Problema: O método list.sort() ordena a lista 'in-place' (modifica a própria lista) e retorna None.
#           Portanto, 'resultado' recebe None, e é isso que será impresso.
lista_a = [3, 1, 2]
lista_a.sort() # A lista é modificada diretamente
print(f"Lista corrigida (a): {lista_a}") # Imprime a lista já ordenada

# b) nomes = ["Ana", "Bruno"]
#    print(nomes[5])
# Problema: IndexError. A lista 'nomes' tem apenas dois elementos (índices 0 e 1).
#           Tentar acessar o índice 5 está fora dos limites da lista.
nomes_b = ["Ana", "Bruno"]
# Reescrevendo de forma segura:
indice_acesso = 5
if indice_acesso < len(nomes_b):
    print(f"Nome no índice {indice_acesso} (b): {nomes_b[indice_acesso]}")
else:
    print(f"Erro (b): Índice {indice_acesso} está fora dos limites da lista de tamanho {len(nomes_b)}.")

# Desafio opcional: Terceiro exemplo de erro (TypeError com concatenação de tipos diferentes)
# c) texto = "Olá" + 123
#    print(texto)
# Problema: TypeError. Não é possível concatenar diretamente uma string com um número inteiro.
#           O número precisa ser convertido para string primeiro.
texto_c = "Olá" + str(123)
print(f"Texto corrigido (c): {texto_c}")

# --- Exercício 14: Projeto integrador curto ---
print("\n--- Exercício 14: Projeto integrador curto ---")

# Dados de exemplo
estudantes_brutos = ["  pedro silva ", "MARIA ALVES  ", "  ana costa"]
notas_alunos = {
    "pedro silva": [6.5, 7.0, 8.0],
    "maria alves": [9.0, 8.5, 9.5],
    "ana costa": [5.0, 6.0, 6.5]
}
consulta_integrador = "maria alves"

# 1. Padronizar nomes de estudantes
nomes_padronizados_integrador = []
for nome_bruto in estudantes_brutos:
    nomes_padronizados_integrador.append(nome_bruto.strip().title())
print("\n--- Relatório Integrado ---")
print(f"Nomes padronizados: {nomes_padronizados_integrador}")

# 2. Filtrar notas aprovadas e calcular média
alunos_aprovados_relatorio = []
notas_aprovadas_relatorio = {}
limite_aprovacao = 7.0

for nome_aluno_dict, notas_list in notas_alunos.items():
    # Padroniza o nome para correspondência com a chave do dicionário (se necessário)
    nome_padronizado_dict = nome_aluno_dict.strip().title()
    media = sum(notas_list) / len(notas_list)
    if media >= limite_aprovacao:
        alunos_aprovados_relatorio.append(nome_padronizado_dict)
        notas_aprovadas_relatorio[nome_padronizado_dict] = media

print(f"Alunos aprovados (média >= {limite_aprovacao}): {alunos_aprovados_relatorio}")

# 3. Verificar a presença de um nome consultado
consulta_integrador_padronizada = consulta_integrador.strip().title()
if consulta_integrador_padronizada in nomes_padronizados_integrador:
    print(f"O estudante '{consulta_integrador_padronizada}' ESTÁ na lista de estudantes.")
else:
    print(f"O estudante '{consulta_integrador_padronizada}' NÃO está na lista de estudantes.")

# 4. Exibir um pequeno relatório final
print("\n--- Detalhes do Relatório Final ---")
print(f"Total de estudantes processados: {len(estudantes_brutos)}")
print(f"Número de estudantes aprovados: {len(alunos_aprovados_relatorio)}")
for aluno, media in notas_aprovadas_relatorio.items():
    print(f"  - {aluno}: Média {media:.2f}")
```
