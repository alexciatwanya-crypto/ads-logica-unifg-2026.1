```python
print("--- Exercício 01: Número positivo, negativo ou zero ---")
numero = int(input("Digite um número inteiro para o Exercício 01: "))

if numero > 0:
    print(f"O número {numero} é positivo.")
elif numero < 0:
    print(f"O número {numero} é negativo.")
else:
    print(f"O número {numero} é zero.")

print("\n--- Exercício 02: Classificação por idade ---")
idade = int(input("Digite a idade da pessoa para o Exercício 02: "))

if idade < 18:
    print(f"Com {idade} anos, a pessoa é menor de idade.")
elif idade >= 60:
    print(f"Com {idade} anos, a pessoa é idosa.")
else:
    print(f"Com {idade} anos, a pessoa é maior de idade.")

print("\n--- Exercício 03: Par ou ímpar de 1 a 20 ---")
for i in range(1, 21):
    if i % 2 == 0:
        print(f"O número {i} é par.")
    else:
        print(f"O número {i} é impar.")

print("\n--- Exercício 04: Senha de acesso ---")
senha_ex4 = ""
while senha_ex4 != "acesso":
    senha_ex4 = input("Exercício 04 - Digite a senha: ").lower()
    if senha_ex4 != "acesso":
        print("Senha incorreta. Tente novamente.")
print("Acesso liberado.")

print("\n--- Exercício 05: Contar números positivos ---")
contador_positivos_ex5 = 0
for i_ex5 in range(1, 11):
    numero_ex5 = float(input(f"Exercício 05 - Digite o {i_ex5}º número: "))
    if numero_ex5 > 0:
        contador_positivos_ex5 += 1

print(f"Foram digitados {contador_positivos_ex5} números positivos.")

print("\n--- Exercício 06: Validação de nota ---")
nota_ex6 = -1

while not (0 <= nota_ex6 <= 10):
    try:
        nota_ex6 = float(input("Exercício 06 - Digite uma nota entre 0 e 10: "))
        if not (0 <= nota_ex6 <= 10):
            print("Valor inválido! A nota deve estar entre 0 e 10.")
    except ValueError:
        print("Entrada inválida! Por favor, digite um número.")

print(f"Nota válida inserida: {nota_ex6}")

print("\n--- Exercício 07: Soma dos números pares ---")
soma_pares_ex7 = 0
for i_ex7 in range(1, 9):
    while True:
        try:
            numero_ex7 = int(input(f"Exercício 07 - Digite o {i_ex7}º número inteiro: "))
            break
        except ValueError:
            print("Entrada inválida! Por favor, digite um número inteiro.")

    if numero_ex7 % 2 == 0:
        soma_pares_ex7 += numero_ex7

print(f"A soma dos números pares digitados é: {soma_pares_ex7}")

print("\n--- Exercício 08: Tabuada ---")
while True:
    try:
        numero_ex8 = int(input("Exercício 08 - Digite um número para ver a tabuada: "))
        break
    except ValueError:
        print("Entrada inválida! Por favor, digite um número inteiro.")

print(f"Tabuada de {numero_ex8}:")
for i_ex8 in range(1, 11):
    resultado_ex8 = numero_ex8 * i_ex8
    print(f"{numero_ex8} x {i_ex8} = {resultado_ex8}")

print("\n--- Exercício 09: Reajuste salarial ---")
while True:
    try:
        salario_original_ex9 = float(input("Exercício 09 - Digite o salário do funcionário: R$ "))
        if salario_original_ex9 < 0:
            print("O salário não pode ser negativo. Tente novamente.")
        else:
            break
    except ValueError:
        print("Entrada inválida! Por favor, digite um valor numérico para o salário.")

percentual_reajuste_ex9 = 0
if salario_original_ex9 <= 1500.00:
    percentual_reajuste_ex9 = 0.15
elif 1500.01 <= salario_original_ex9 <= 3000.00:
    percentual_reajuste_ex9 = 0.10
else:
    percentual_reajuste_ex9 = 0.05

novo_salario_ex9 = salario_original_ex9 * (1 + percentual_reajuste_ex9)

print(f"Salário Original: R$ {salario_original_ex9:.2f}")
print(f"Percentual Aplicado: {percentual_reajuste_ex9 * 100:.0f}%")
print(f"Novo Salário: R$ {novo_salario_ex9:.2f}")

print("\n--- Exercício 10: Menu simples ---")
opcao_ex10 = -1

while opcao_ex10 != 0:
    print("\n--- Menu (Exercício 10) ---")
    print("1 - Somar")
    print("2 - Subtrair")
    print("0 - Sair")

    try:
        opcao_ex10 = int(input("Exercício 10 - Escolha uma opção: "))

        if opcao_ex10 == 1:
            num1_ex10 = float(input("Exercício 10 - Digite o primeiro número para somar: "))
            num2_ex10 = float(input("Exercício 10 - Digite o segundo número para somar: "))
            print(f"Resultado da soma: {num1_ex10 + num2_ex10}")
        elif opcao_ex10 == 2:
            num1_ex10 = float(input("Exercício 10 - Digite o primeiro número para subtrair: "))
            num2_ex10 = float(input("Exercício 10 - Digite o segundo número para subtrair: "))
            print(f"Resultado da subtração: {num1_ex10 - num2_ex10}")
        elif opcao_ex10 == 0:
            print("Saindo do programa do Exercício 10.")
        else:
            print("Opção inválida! Por favor, escolha 0, 1 ou 2.")
    except ValueError:
        print("Entrada inválida! Por favor, digite um número correspondente à opção.")

print("\n--- Exercício 11: Classificação da turma ---")
while True:
    try:
        quantidade_alunos_ex11 = int(input("Exercício 11 - Digite a quantidade de alunos na turma: "))
        if quantidade_alunos_ex11 <= 0:
            print("A quantidade de alunos deve ser um número positivo.")
        else:
            break
    except ValueError:
        print("Entrada inválida! Por favor, digite um número inteiro.")


aprovados_ex11 = 0
recuperacao_ex11 = 0
reprovados_ex11 = 0

for i_ex11 in range(1, quantidade_alunos_ex11 + 1):
    while True:
        try:
            media_ex11 = float(input(f"Exercício 11 - Digite a média do aluno {i_ex11}: "))
            if 0 <= media_ex11 <= 10:
                break
            else:
                print("A média deve ser entre 0 e 10.")
        except ValueError:
            print("Entrada inválida! Por favor, digite um número para a média.")

    if media_ex11 >= 7:
        aprovados_ex11 += 1
    elif media_ex11 >= 4:
        recuperacao_ex11 += 1
    else:
        reprovados_ex11 += 1

print(f"\nTotal de Alunos Aprovados: {aprovados_ex11}")
print(f"Total de Alunos em Recuperação: {recuperacao_ex11}")
print(f"Total de Alunos Reprovados: {reprovados_ex11}")

print("\n--- Exercício 12: Classificação de compras ---")
continuar_ex12 = 's'
while continuar_ex12.lower() == 's':
    while True:
        try:
            valor_compra_ex12 = float(input("Exercício 12 - Digite o valor da compra: R$ "))
            if valor_compra_ex12 < 0:
                print("O valor da compra não pode ser negativo. Tente novamente.")
            else:
                break
        except ValueError:
            print("Entrada inválida! Por favor, digite um valor numérico.")

    if valor_compra_ex12 <= 100.00:
        print(f"Compra de R$ {valor_compra_ex12:.2f}: Sem desconto.")
    elif 100.01 <= valor_compra_ex12 <= 500.00:
        print(f"Compra de R$ {valor_compra_ex12:.2f}: Desconto básico.")
    else:
        print(f"Compra de R$ {valor_compra_ex12:.2f}: Desconto especial.")

    continuar_ex12 = input("Exercício 12 - Deseja classificar outra compra? (s/n): ")

print("Classificação de compras encerrada.")

print("\n--- Exercício 13: Depuração de loop infinito ---")
contador_ex13 = 10
while contador_ex13 >= 1:
    print(contador_ex13)
    contador_ex13 -= 1

print("Contagem finalizada.")

print("\n--- Exercício 14: Reescrita de while para for ---")
print("--- Versão com while ---")
contador_while_ex14 = 1
while contador_while_ex14 <= 5:
    print(contador_while_ex14)
    contador_while_ex14 += 1

print("\n--- Versão com for ---")
for contador_for_ex14 in range(1, 6):
    print(contador_for_ex14)

print("\n--- Exercício 15: Maior nota da sequência ---")
maior_nota_ex15 = -1.0

for i_ex15 in range(1, 6):
    while True:
        try:
            nota_ex15 = float(input(f"Exercício 15 - Digite a {i_ex15}ª nota (0-10): "))
            if 0 <= nota_ex15 <= 10:
                break
            else:
                print("Nota inválida! Digite um valor entre 0 e 10.")
        except ValueError:
            print("Entrada inválida! Por favor, digite um número.")

    if nota_ex15 > maior_nota_ex15:
        maior_nota_ex15 = nota_ex15

print(f"A maior nota da sequência é: {maior_nota_ex15}")

print("\n--- Exercício 16: Questionário com sentinela ---")
respostas_sim_ex16 = 0
resposta_ex16 = ""

print("Questionário: Digite 'S' para sim, 'N' para não, ou 'FIM' para encerrar.")

while resposta_ex16.upper() != "FIM":
    resposta_ex16 = input("Sua resposta (S/N/FIM): ").upper()

    if resposta_ex16 == "S":
        respostas_sim_ex16 += 1
    elif resposta_ex16 == "N":
        pass
    elif resposta_ex16 == "FIM":
        print("Questionário encerrado.")
    else:
        print("Resposta inválida! Por favor, digite 'S', 'N' ou 'FIM'.")

print(f"\nTotal de respostas 'S' (sim): {respostas_sim_ex16}")

print("\n--- Desafio Extra: Sistema simples de lançamento de desempenho ---")
estudantes_extra = []

while True:
    nome_extra = input("Desafio Extra - Digite o nome do estudante (ou 'FIM' para encerrar): ")
    if nome_extra.upper() == 'FIM':
        break

    while True:
        try:
            acertos_extra = int(input(f"Desafio Extra - Digite a quantidade de acertos de {nome_extra} (0 a 20): "))
            if 0 <= acertos_extra <= 20:
                break
            else:
                print("Acertos inválidos! Digite um número entre 0 e 20.")
        except ValueError:
            print("Entrada inválida! Por favor, digite um número inteiro.")

    percentual_acertos_extra = (acertos_extra / 20) * 100

    desempenho_extra = ""
    if percentual_acertos_extra >= 90:
        desempenho_extra = "Excelente"
    elif percentual_acertos_extra >= 70:
        desempenho_extra = "Satisfatório"
    elif percentual_acertos_extra >= 50:
        desempenho_extra = "Em atenção"
    else:
        desempenho_extra = "Crítico"

    estudantes_extra.append({"nome": nome_extra, "acertos": acertos_extra, "percentual": percentual_acertos_extra, "desempenho": desempenho_extra})
    print(f"Desempenho de {nome_extra}: {percentual_acertos_extra:.2f}% ({desempenho_extra})\n")

if estudantes_extra:
    total_percentual_extra = sum(s['percentual'] for s in estudantes_extra)
    media_turma_extra = total_percentual_extra / len(estudantes_extra)
    print("--- Relatório Final (Desafio Extra) ---")
    for s_extra in estudantes_extra:
        print(f"Estudante: {s_extra['nome']}, Acertos: {s_extra['acertos']}/20, Percentual: {s_extra['percentual']:.2f}%, Desempenho: {s_extra['desempenho']}")
    print(f"\nMédia percentual da turma: {media_turma_extra:.2f}%")
else:
    print("Nenhum estudante foi cadastrado.")
```
