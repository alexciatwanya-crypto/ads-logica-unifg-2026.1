```python
# Exercicios praticos

# 1) Saudacao modular
def saudacao(nome):
    return f"Bem-vindo(a), {nome}!"

print("--- Exercicio 1 ---")
print(saudacao("Alexcia"))
print(saudacao("Bruno"))
print(saudacao("Anderson"))
print("\n")

# 2) Soma com parametros
def somar(a, b):
    return a + b

print("--- Exercicio 2 ---")
print(f"Soma de 5 e 3: {somar(5, 3)}") # Inteiros
print(f"Soma de 10 e 0: {somar(10, 0)}") # Com zero
print(f"Soma de -7 e 2: {somar(-7, 2)}") # Com numero negativo
print("\n")

# 3) Media de aluno
def calcular_media(n1, n2):
    return (n1 + n2) / 2

print("--- Exercicio 3 ---")
try:
    nota1 = float(input("Digite a primeira nota: "))
    nota2 = float(input("Digite a segunda nota: "))
    media = calcular_media(nota1, nota2)
    print(f"A media das notas e: {media:.1f}")
except ValueError:
    print("Entrada invalida. Por favor, digite numeros.")
print("\n")

# 4) Situacao do aluno
def verificar_situacao(media):
    if media >= 7:
        return "Aprovado"
    else:
        return "Reprovado"

print("--- Exercicio 4 ---")
# Reutilizando a funcao calcular_media do exercicio anterior
try:
    nota1_ex4 = float(input("Digite a primeira nota para Exercicio 4: "))
    nota2_ex4 = float(input("Digite a segunda nota para Exercicio 4: "))
    media_ex4 = calcular_media(nota1_ex4, nota2_ex4)
    situacao_ex4 = verificar_situacao(media_ex4)
    print(f"Media: {media_ex4:.1f}, Situacao: {situacao_ex4}")
except ValueError:
    print("Entrada invalida. Por favor, digite numeros.")
print("\n")

# 5) Boletim simples com modularizacao
def ler_notas():
    while True:
        try:
            n1 = float(input("Digite a primeira nota: "))
            n2 = float(input("Digite a segunda nota: "))
            return n1, n2
        except ValueError:
            print("Entrada invalida. Por favor, digite numeros.")

# calcular_media e verificar_situacao ja definidas

def exibir_resultado(nome, media, situacao):
    print(f"Nome do Aluno: {nome}")
    print(f"Media Final: {media:.1f}")
    print(f"Situacao: {situacao}")

print("--- Exercicio 5 ---")
aluno_nome = input("Digite o nome do aluno: ")
notas_ex5_n1, notas_ex5_n2 = ler_notas()
media_ex5 = calcular_media(notas_ex5_n1, notas_ex5_n2)
situacao_ex5 = verificar_situacao(media_ex5)
exibir_resultado(aluno_nome, media_ex5, situacao_ex5)
print("\n")


# 6) Escopo basico na pratica
x = 10

def teste():
    y = 5
    return x + y

print("--- Exercicio 6 ---")
print(f"Resultado da funcao teste(): {teste()}")

# 1. Identifique qual variavel e global.
print("Variavel global: 'x' (declarada fora de qualquer funcao).")

# 2. Identifique qual variavel e local.
print("Variavel local: 'y' (declarada dentro da funcao 'teste').")

# 3. Explique o que acontece se tentarmos usar `y` fora da funcao.
print("Se tentarmos usar 'y' fora da funcao 'teste', ocorrera um 'NameError', pois 'y' so existe dentro do escopo da funcao onde foi definida.")
# Exemplo comentado de erro:
# try:
#     print(y)
# except NameError as e:
#     print(f"Erro ao tentar acessar 'y' fora da funcao: {e}")

# 4. Crie um exemplo parecido feito por voce.
var_global = "Eu sou global"

def minha_funcao():
    var_local = "Eu sou local"
    print(f"Dentro da funcao: {var_global} e {var_local}")

minha_funcao()
print(f"Fora da funcao: {var_global}")
# print(var_local) # Isso causaria um NameError
print("\n")

# 7) Refatorando codigo monolitico
print("--- Exercicio 7 ---")
def calcular_media_refatorado(n1, n2):
    return (n1 + n2) / 2

def verificar_situacao_refatorado(media):
    if media >= 7:
        return "Aprovado"
    else:
        return "Reprovado"

# Exemplo de uso:
n1_ex7 = 8
n2_ex7 = 6
media_ex7 = calcular_media_refatorado(n1_ex7, n2_ex7)
situacao_ex7 = verificar_situacao_refatorado(media_ex7)
print(f"Notas: {n1_ex7}, {n2_ex7} -> Media: {media_ex7}, Situacao: {situacao_ex7}")

n1_ex7_b = 5
n2_ex7_b = 6
media_ex7_b = calcular_media_refatorado(n1_ex7_b, n2_ex7_b)
situacao_ex7_b = verificar_situacao_refatorado(media_ex7_b)
print(f"Notas: {n1_ex7_b}, {n2_ex7_b} -> Media: {media_ex7_b}, Situacao: {situacao_ex7_b}")
print("\n")

# 8) Depuracao de erro sintatico
print("--- Exercicio 8 ---")
# Codigo com erro original:
# def saudacao(nome)
#     print("Ola,", nome)

# Correcao:
def saudacao_corrigida(nome):
    print("Ola,", nome)

saudacao_corrigida("Maria")

print("O erro era a falta dos dois pontos (:) apos a definicao da funcao 'saudacao(nome)'.")
print("Sem os dois pontos, o interpretador Python nao consegue reconhecer o inicio do bloco da funcao, resultando em um 'SyntaxError' e impedindo a execucao do programa.")
print("\n")

# 9) Depuracao de erro de execucao
print("--- Exercicio 9 ---")
def dividir(a, b):
    if b == 0:
        return "Erro: Divisao por zero nao e permitida."
    return a / b

print(f"Divisao de 10 por 2: {dividir(10, 2)}")
print(f"Divisao de 10 por 0: {dividir(10, 0)}")

print("1. O erro que acontece e um 'ZeroDivisionError'.")
print("2. Ele ocorre porque qualquer numero dividido por zero e matematicamente indefinido, e em programacao, essa operacao nao e permitida.")
print("3. O programa foi reescrito para incluir uma verificacao se 'b' e zero antes de realizar a divisao, retornando uma mensagem amigavel neste caso.")
print("\n")

# 10) Teste de mesa + caso de teste
def calcular_total(preco, quantidade):
    subtotal = preco * quantidade
    desconto = subtotal * 0.1 # 10% de desconto
    total = subtotal - desconto
    return total

print("--- Exercicio 10 ---")
print("Parte A - Teste de mesa para calcular_total(50, 2)")
print("preco = 50")
print("quantidade = 2")
print("subtotal = preco * quantidade = 50 * 2 = 100")
print("desconto = subtotal * 0.1 = 100 * 0.1 = 10")
print("total = subtotal - desconto = 100 - 10 = 90")
print(f"Resultado calculado: {calcular_total(50, 2)}")
print("\n")

print("Parte B - Casos de teste para calcular_total")
# 2 casos normais
print(f"Caso normal 1 (10, 3): {calcular_total(10, 3)}") # Esperado: 10*3 - (10*3)*0.1 = 30 - 3 = 27
print(f"Caso normal 2 (100, 1): {calcular_total(100, 1)}") # Esperado: 100*1 - (100*1)*0.1 = 100 - 10 = 90

# 1 caso limitrofe (ex: quantidade minima, preco minimo, desconto zero)
print(f"Caso limitrofe (preco 0, quantidade 5): {calcular_total(0, 5)}") # Esperado: 0

# 1 caso extremo (ex: numeros grandes)
print(f"Caso extremo (1000000, 10): {calcular_total(1000000, 10)}") # Esperado: 10M - 1M = 9M
print("\n")

# Desafio final opcional
print("--- Desafio Final ---")
def ler_dados_aluno():
    nome = input("Digite o nome do aluno para o desafio final: ")
    while True:
        try:
            n1 = float(input("Digite a primeira nota: "))
            n2 = float(input("Digite a segunda nota: "))
            return nome, n1, n2
        except ValueError:
            print("Entrada invalida. Por favor, digite numeros para as notas.")

def calcular_media_desafio(n1, n2):
    return (n1 + n2) / 2

def verificar_situacao_desafio(media):
    if media >= 7:
        return "Aprovado"
    else:
        return "Reprovado"

def exibir_relatorio_final(nome, media, situacao):
    print("\n--- Relatorio Final do Aluno ---")
    print(f"Aluno: {nome}")
    print(f"Media: {media:.1f}")
    print(f"Situacao: {situacao}")
    print("-------------------------------")


nome_desafio, n1_desafio, n2_desafio = ler_dados_aluno()
media_desafio = calcular_media_desafio(n1_desafio, n2_desafio)
situacao_desafio = verificar_situacao_desafio(media_desafio)
exibir_relatorio_final(nome_desafio, media_desafio, situacao_desafio)
print("\n")

print("--- Fim dos Exercicios --- ")
```
